## Ques 1 = What does RGBA stand for?

Answer - RGBA(Red-Green-Blue-Alpha)

The RGB color model is extended in this specification to include "alpha" to allow specification of the opacity of a color.

- An RGBA value is a tuple of 4 integers each ranging from 0 to 255.The four interger correspond to the amount of red,green,blue and alpha (transparency) of the color.
